```python
data1=pd.read_csv('employee_data.csv')
data2=pd.read_csv('manager_data.csv')
df=pd.merge(data1,data2,on='employee_name',how='outer')
```

```python
df=df.drop(['gender'],axis=1)
```

```python
df
```

```python
df.rename(columns = {'employee_name':'name', 'leave_dates':'fromdate','leave_duration':'todate','leave_type':'leave_type', 'manager_name':'reporter','team_name':'team_name'}, inplace = True)
```

```python
df
```

```python
df['fromdate']=pd.to_datetime(df['fromdate'])
```

```python
df['fromdate'] = df['fromdate'].astype(str)
df['todate'] = df['todate'].astype(str)
```

```python
df1=df.to_json(orient="records")
```

```python
df1= df1.replace("'","")
```

```python
df1
```

```python
df2=json.loads(df1)
```

```python
url = "http://localhost:4747/post"
headers = {
    'Content-Type': "application/json"
    }
for i in range(0,len(df2)):
    df2[i]['fromdate']=df2[i]['fromdate']+'T00:00:00Z'
    resp = requests.request("POST", url,headers=headers, json=df2[i])
    json.loads(resp.text)
```

```python

CREATE TABLE Notifications (
    reporting_manager VARCHAR(100) NOT NULL,
    leave_id INT PRIMARY KEY,
    approved BOOLEAN DEFAULT FALSE
);

CREATE FUNCTION auditlogfunc() RETURNS TRIGGER AS $example_table$
   BEGIN
      INSERT INTO Notifications(reporting_manager, leave_id,approved) VALUES (new.reporter, new.id,'False');
      RETURN NEW;
   END;
$example_table$ LANGUAGE plpgsql;

CREATE TRIGGER example_trigger AFTER INSERT ON employees
FOR EACH ROW EXECUTE PROCEDURE auditlogfunc();


drop TRIGGER example_trigger on employees;

drop FUNCTION auditlogfunc;
select * from notifications;
```
